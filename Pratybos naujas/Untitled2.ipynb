```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file (First sheet only)
file_path = r"slope_offset_median_mean_results.xlsx"  # Update your file path
xls = pd.ExcelFile(file_path)
sheet_name = xls.sheet_names[0]  # Get the first sheet name

# Read only the first two columns
df = pd.read_excel(xls, sheet_name=sheet_name, usecols=[0, 1])

# Rename columns for clarity
df.columns = ["files", "median_slope"]

# Extract group prefix (assuming groups are the first few letters of the subject name)
df["Group"] = df["files"].str.extract(r"([A-Za-z]+)")  # Extracts letters as group

# Create a grouped violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x="Group", y="median_slope", data=df, inner="point", palette="Set2")

# Format the plot
plt.title(f"Violin Plot by Group - {sheet_name}", fontsize=14)
plt.xlabel("Group", fontsize=12)
plt.ylabel("Median Slope", fontsize=12)
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels for better readability
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Save the plot
output_path = f"plots/{sheet_name}_grouped_violin_plot.png"
plt.savefig(output_path, dpi=300, bbox_inches="tight")
plt.show()

print(f"Plot saved as {output_path}")
```
