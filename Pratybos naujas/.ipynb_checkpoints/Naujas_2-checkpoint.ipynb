```python
import os
import mne
import pandas as pd 
import numpy as np

# Define directories 
load_dir = 'E:\\Analysis\\RAW_resempled_setschanels\\NCG' 
save_dir = 'E:\\Analysis\\PSDs and freq'

# List all .set files in the directory 
files = [f for f in os.listdir(load_dir) if f.endswith('.set')]

# Check if files exist
if not files: 
    raise ValueError(f'No .set files found in directory: {load_dir}') 
else: 
    print(f'{len(files)} .set files found in directory: {load_dir}')

# Initialize lists to store data
all_participant_ids = []
all_frequencies = []
all_psds = []

for isb, file in enumerate(files):
    try:
        # Load the EEG data
        eeg_file = os.path.join(load_dir, file)
        raw = mne.io.read_raw_eeglab(eeg_file, preload=True)
        participant_id = os.path.splitext(file)[0]
        
        # Compute PSD using Welch's method
        psds, freqs = mne.time_frequency.psd_welch.multitaper(raw, fmin=1, fmax=45, n_fft=2048)
        psds = np.mean(psds, axis=0)  # Average across channels
        
        # Store data
        all_participant_ids.append(participant_id)
        all_frequencies.append(freqs)
        all_psds.append(psds)
        
        print(f'Processed file: {file}')
    except Exception as e:
        print(f'Error processing file: {file}')
        print(f'Error message: {e}')
```

```python
import os
import mne
import pandas as pd
import numpy as np

# Define directories
aft_dir = 'E:\\Analysis\\re-referenced\\NCF'
save_dir = 'E:\\Analysis\\PSDs and freq'

# Ensure save_dir exists, create if not
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    print(f'Created directory: {save_dir}')

# List all .set files in the directory
files = [f for f in os.listdir(aft_dir) if f.endswith('.set')]

# Check if files exist
if not files:
    raise ValueError(f'No .set files found in directory: {aft_dir}')
else:
    print(f'{len(files)} .set files found in directory: {aft_dir}')

# Initialize lists to store data
all_participant_ids = []
all_frequencies = []
all_psds = []

for isb, file in enumerate(files):
    try:
        # Load the EEG data
        eeg_file = os.path.join(aft_dir, file)
        raw = mne.io.read_raw_eeglab(eeg_file, preload=True)
        participant_id = os.path.splitext(file)[0]
        
        # Compute PSD using Welch's method
        psds, freqs = mne.time_frequency.psd_welch(raw, fmin=1, fmax=45, n_fft=2048)
        psds = np.mean(psds, axis=0)  # Average across channels
        
        # Store data
        all_participant_ids.append(participant_id)
        all_frequencies.append(freqs)
        all_psds.append(psds)
        
        print(f'Processed file: {file}')
    except Exception as e:
        print(f'Error processing file: {file}')
        print(f'Error message: {e}')

# Save data to an Excel file
data = {
    'Participant_ID': all_participant_ids,
    'Frequencies': [list(freq) for freq in all_frequencies],
    'Power_Spectra': [list(psd) for psd in all_psds]
}
df = pd.DataFrame(data)
output_path = os.path.join(save_dir, 'psd_freq_data.xlsx')
df.to_excel(output_path, index=False, engine='openpyxl')

print(f'PSD data saved to {output_path}')
```
